```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
gc=pd.read_csv('gender_submission.csv')
print(df.isnull().sum()) # to see total no of 0s in a column
```

```python
df.drop('Name',axis=1,inplace=True)
df.drop('Ticket',axis=1,inplace=True)
df.drop('Cabin',axis=1,inplace=True)
df.drop('PassengerId',axis=1,inplace=True)
df.dropna(inplace = True) #to drop all the rows with nan val
print(df)

test.drop('Name',axis=1,inplace=True)
test.drop('Ticket',axis=1,inplace=True)
test.drop('Cabin',axis=1,inplace=True)
test.drop('PassengerId',axis=1,inplace=True)
print(test)
```

```python
print(test.isnull().sum())
test.info()
```

```python
gc.drop('PassengerId',axis=1,inplace=True)
print(gc)
```

```python
X=df.drop('Survived',axis=1)
y=df['Survived']
print(X)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df[['Age']] = imputer.fit_transform(df[['Age']])

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
test[['Age']] = imputer.fit_transform(test[['Age']])
test[['Fare']] = imputer.fit_transform(test[['Fare']])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['Sex']= le.fit_transform(X['Sex'])
unique_species = X['Sex'].unique()
print(X)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
test['Sex']= le.fit_transform(test['Sex'])
unique_species = test['Sex'].unique()
print(test)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
X=ct.fit_transform(X)
print(X)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [6])], remainder='passthrough')
test=ct.fit_transform(test)
print(test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X,y)     #only X and y if training on whole dataset
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X,y)   #only X and y if training on whole dataset
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X, y)    #only X and y if training on whole dataset
#Maximum
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X,y)    #only X and y if training on whole dataset
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X,y)   #only X and y if training on whole dataset
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=15,criterion='entropy',random_state=0)
classifier.fit(X,y)  #only X and y if training on whole dataset
```
